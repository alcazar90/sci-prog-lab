### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [X] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [X] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [X] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [X] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [X] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como guía para desarrollar este punto.

````python
# Promedio de notas
df_grades["GPA"] = df_grades[
    ["math score", "reading score", "writing score"]
].mean(axis=1)

# Mapeo
df_grades["simple parental level of education"] = df_grades[
    "parental level of education"
].map({
    "some high school": "school",
    "some college": "school",
    "high school": "school",
    "bachelor's degree": "college",
    "associate's degree": "college",
    "master's degree": "postgraduate"
})

resumen = df_grades.groupby([
    "race/ethnicity", "simple parental level of education"
], as_index=False).agg({
    "math score": "mean",
    "reading score": "mean",
    "writing score": "mean",
    "GPA": "mean",
    "names": [len, lambda x: f"{round(len(x) / df_grades.shape[0] * 100, 2)} %"]
}).droplevel(1, axis=1).round(2)

resumen.columns = list(resumen.columns[:-2]) + ["count", "percentage"]

resumen.loc[resumen.duplicated("race/ethnicity"), "race/ethnicity"] = ""
display(resumen)
````

````python
# Cargar other_grades
df_other = pd.read_csv(
      "https://raw.githubusercontent.com/alcazar90/sci-prog-lab/main/data/other_grades.csv"
)

df_other.head()
````

````python
# Unir df_grades con other_grades.csv usando outer join
df_outer_join = pd.merge(
    df_grades,
    df_other,
    how="outer"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_outer_join: {df_outer_join.shape}"
)
print(
    #f"\nNulos df_grades: \n{df_grades.isna().sum()}\n"
    #f"\nNulos df_other: \n{df_other.isna().sum()}\n"
    f"\nNulos df_outer_join: \n{df_outer_join.isna().sum()}"
)
display(df_outer_join.head())
````

**Explicación del resultado**

- El DataFrame resultante contiene las filas presentes en los dos DataFrame originales; Aquellas filas comunes en ambos DataFrame originales se mantienen, y además se agregan las filas adicionales que, en este caso, contiene `df_other`.
- Debido a lo anterior, `df_outer_join` contiene _1000 filas_, y 13 columnas, que corresponde a las 11 columnas de `df_grades` más las 2 columnas adicionales de `df_other` (i.e. `science score` y `history score`).
- La columna `names` es común a ambos DataFrame, por lo que se utiliza como clave de unión, por lo que la cantidad resultante de columnas es 11 + 3 - 1 = 13.
- La cantidad de nulos resultantes corresponde a las filas "adicionales" presentes en `df_other`, las cuales al no poder asociarse a registros de `df_grades`, resultan en valores nulos para todas las columnas presentes en `df_grades` (a excepción de `name`, ya que esta columna sí está presente en `df_other`)

````python
df_other["names"].nunique(), df_outer_join["names"].nunique()
````

- Se comprueba que `df_other` no tiene nombres duplicados, con lo que se comprueba que las filas adicionales corresponden a alumnos que no se encontraban en `df_grades`, resultando así las columnas correspondientes a ese DataFrame en valores nulos al momento de hacer el merge.

````python
# Unir df_grades con other_grades.csv usando left join
df_left_join = pd.merge(
    df_grades,
    df_other,
    how="left"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_left_join: {df_left_join.shape}"
)
print(
    f"\nNulos df_left_join: \n{df_left_join.isna().sum()}"
)
df_left_join.head()
````

**Explicación**

- El DataFrame que se obtiene contiene solo el número de filas
originales en `df_grades`.
- Sin embargo, se modifica la información a nivel de columnas, anexando
las columnas adicionales que se encuentran en `other_grades` (i.e. `science score` y `history score`). Por lo tanto, pasamos de 11 + 2 = 13 columnas.
- Leémos `left` como el dataframe en la primera posición de la función
`merge(LEFT DATAFRAME, RIGHT DATAFRAME, ...)`.
- Al utilizar el argumento `how='left'` dentro de la función `merge()`, 
estamos garantizando que mantendremos solo las llaves encontradas en
el DataFrame de la izquierda (`nombres`). Esto significa que solo se 
agregaran las columnas donde el nombre se encuentre en la izquierda, cualquier
nombre que este en el DataFrame `other_grades`, pero no en `df_grades`, su información no será anexada.
- Este método es relevante cuando queremos incrementar el número de información
procedente de otras fuentes sobre un dominio de interes. Nos interesa
saber más de los alumnos que tienen todas las notas  `math score`, `reading score`, y `writing score`. Si tenemos notas de otras asignaturas, genial, pero
no queremos seguir alumnos que no tengan al menos las 3 primeras notas. Este
sería un caso de uso para el `left join`.
- La llave al no ser especificada en `merge()` se determina por las columnas
que tienen igual nombre en ambos DataFrames (i.e. `nombre`).



````python
# Unir df_grades con other_grades.csv usando right join
df_right_join = pd.merge(
    df_grades,
    df_other,
    how="right"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_right_join: {df_right_join.shape}"
)
print(
    f"\nNulos df_right_join: \n{df_right_join.isna().sum()}"
)
df_right_join.head()
````

**Explicación**

- Similar al razonamiento de `how='left'`, pero con el DataFrame del segundo
argumento en `merge()`, el resultado de la operación contiene solo el número de filas originales de `other_grades`.
- Sin embargo, se modifica la información a nivel de columnas, anexando las columnas adicionales que se encuentran en `df_grades` (i.e. `gender` a `simple parental level of education``). Por lo tanto, pasamos de 3 + 10 = 13 columnas.
- Leémos `right` como el DataFrame en la segunda posición de la función
`merge(LEFT DATAFRAME, RIGHT DATAFRAME, ...)`.
- Al utilizar el argumento `how='right'` dentro de la función merge(), estamos garantizando que mantendremos solo las llaves encontradas en el DataFrame de la izquierda (`nombres`). Esto significa que solo se agregará las columnas donde el `nombre` se encuentre en la derecha, cualquier nombre que este en el DataFrame `df_grades`, pero no en `other_grades`, su información no será anexada.
- La llave al no ser especificada en `merge()` se determina por las columnas
que tienen igual nombre en ambos DataFrames (i.e. `nombre`).

````python
# Unir df_grades con other_grades.csv usando inner join
df_inner_join = pd.merge(
    df_grades,
    df_other,
    how="inner"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_inner_join: {df_inner_join.shape}"
)
print(
    f"\nNulos df_inner_join: \n{df_inner_join.isna().sum()}"
)
df_inner_join.head()
````

**Explicación**

- Se anexa información adicional para toda observación que contenga una
llave en el DataFrame `df_grades` Y `other_grades`. 
- Es una operación que se fija en la información en la intersección de las
llaves de ambos DataFrame.
- La llave al no ser especificada en `merge()` se determina por las columnas que tienen igual nombre en ambos DataFrames (i.e. nombre).
- El DataFrame que da como resultado contiene 13 columnas, las 11 que se
encuentran en `df_grades` y dos nuevas columnas adicionales aportadas por
`other_grades`.
- Respecto al número de filas, tenemos un total de 875 observaciones, esto
quiere decir que hay solo 875 `nombres` compartidos en ambos DataFrames.

````python
# Transformar a númericas las columnas race/ethnicity y simple parental level of education
df_coord_para = resumen.copy()
race=np.repeat(df_coord_para['race/ethnicity'].values, 3)
df_coord_para['race/ethnicity'] = race[race != '']
df_coord_para["race/ethnicity"] = df_coord_para["race/ethnicity"].replace({"group A": 0,
                                                                           "group B": 1,
                                                                           "group C": 2,
                                                                           "group D": 3,
                                                                           "group E": 4})


df_coord_para["simple parental level of education"] = df_coord_para["simple parental level of education"].replace({
   "school": 0,
   "college": 1,
   "postgraduate": 2})
````

````python
fig = px.parallel_coordinates(
    df_coord_para[['race/ethnicity', 'math score', 'reading score', 'writing score',
                   'GPA', 'count', 'simple parental level of education']],
    color="simple parental level of education",
    color_continuous_scale=px.colors.diverging.Tealrose,
    color_continuous_midpoint=1,
    title="Ruta de la etnicidad del estudiante por sus notas hasta el nivel educacional de sus padres.<br>"
          "<sup>Race/Ethnicity group ->  A0, B1, C2, D3, E4 |  Education -> school (0), college (1), postgraduate (2).</sup>"
    )

fig.show()
````

**Segunda visualización:** continuamos explicando y analizando la información
del desempeño de los alumnos entre las variables de grupo etnico al que
pertenecen y el nivel de educación de sus padres.

````python
# Transformar a númericas las columnas race/ethnicity y simple parental level of education
df_heatmap = resumen.copy()
race=np.repeat(df_heatmap['race/ethnicity'].values, 3)
df_heatmap['race/ethnicity'] = race[race != '']
df_heatmap.sample(n=3)
````

````python
fig = px.density_heatmap(df_heatmap, 
                         x="race/ethnicity", 
                         y="simple parental level of education",
                         z="GPA", text_auto=True,
                         title="Distribución bivariada del GPA(math, reading, writing)"
                               "<br><sup>     Por nivel educacional de los padres y grupo etnico"
                               " al que pertenecen los estudiantes</sup>",
                         color_continuous_scale="viridis")

fig.show()
````

cada grupo y facilita la comparación. Se destaca de la visualización los grupos más bajos (i.e. school A-B), 
homogeneos (i.e. college B-C-D), o con promedios más altos (i.e. postgraduate-E).

Esta visualización promueve el contraste de patrones a un alto nivel y nos 
facilita las comparaciones que sean relativamente pequeñas entre grupos. Una forma
de alivianar este problema es agregando etiquetas con las anotaciones del
`GPA` para cada uno de los 15 grupos.
