### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [X] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [X] Hacer una simplificaci칩n a trav칠s de un mapeo (investigar el m칠todo `map()`) de la variable `parental level of education`  seg칰n la siguiente conversi칩n: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [X] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [X] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [X] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como gu칤a para desarrollar este punto.

````python
# Promedio de notas
df_grades["GPA"] = df_grades[
    ["math score", "reading score", "writing score"]
].mean(axis=1)

# Mapeo
df_grades["simple parental level of education"] = df_grades[
    "parental level of education"
].map({
    "some high school": "school",
    "some college": "school",
    "high school": "school",
    "bachelor's degree": "college",
    "associate's degree": "college",
    "master's degree": "postgraduate"
})

resumen = df_grades.groupby([
    "race/ethnicity", "simple parental level of education"
], as_index=False).agg({
    "math score": "mean",
    "reading score": "mean",
    "writing score": "mean",
    "GPA": "mean",
    "names": [len, lambda x: f"{round(len(x) / df_grades.shape[0] * 100, 2)} %"]
}).droplevel(1, axis=1).round(2)

resumen.columns = list(resumen.columns[:-2]) + ["count", "percentage"]

resumen.loc[resumen.duplicated("race/ethnicity"), "race/ethnicity"] = ""
display(resumen)
````

````python
# Cargar other_grades
df_other = pd.read_csv(
      "https://raw.githubusercontent.com/alcazar90/sci-prog-lab/main/data/other_grades.csv"
)

df_other.head()
````

````python
# Unir df_grades con other_grades.csv usando outer join
df_outer_join = pd.merge(
    df_grades,
    df_other,
    how="outer"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_outer_join: {df_outer_join.shape}"
)
print(
    #f"\nNulos df_grades: \n{df_grades.isna().sum()}\n"
    #f"\nNulos df_other: \n{df_other.isna().sum()}\n"
    f"\nNulos df_outer_join: \n{df_outer_join.isna().sum()}"
)
display(df_outer_join.head())
````

**Explicaci칩n del resultado**

- El DataFrame resultante contiene las filas presentes en los dos DataFrame originales; Aquellas filas comunes en ambos DataFrame originales se mantienen, y adem치s se agregan las filas adicionales que, en este caso, contiene `df_other`.
- Debido a lo anterior, `df_outer_join` contiene _1000 filas_, y 13 columnas, que corresponde a las 11 columnas de `df_grades` m치s las 2 columnas adicionales de `df_other` (i.e. `science score` y `history score`).
- La columna `names` es com칰n a ambos DataFrame, por lo que se utiliza como clave de uni칩n, por lo que la cantidad resultante de columnas es 11 + 3 - 1 = 13.
- La cantidad de nulos resultantes corresponde a las filas "adicionales" presentes en `df_other`, las cuales al no poder asociarse a registros de `df_grades`, resultan en valores nulos para todas las columnas presentes en `df_grades` (a excepci칩n de `name`, ya que esta columna s칤 est치 presente en `df_other`)

````python
df_other["names"].nunique(), df_outer_join["names"].nunique()
````

- Se comprueba que `df_other` no tiene nombres duplicados, con lo que se comprueba que las filas adicionales corresponden a alumnos que no se encontraban en `df_grades`, resultando as칤 las columnas correspondientes a ese DataFrame en valores nulos al momento de hacer el merge.

````python
# Unir df_grades con other_grades.csv usando left join
df_left_join = pd.merge(
    df_grades,
    df_other,
    how="left"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_left_join: {df_left_join.shape}"
)
print(
    f"\nNulos df_left_join: \n{df_left_join.isna().sum()}"
)
df_left_join.head()
````

**Explicaci칩n**

- El DataFrame que se obtiene contiene solo el n칰mero de filas
originales en `df_grades`.
- Sin embargo, se modifica la informaci칩n a nivel de columnas, anexando
las columnas adicionales que se encuentran en `other_grades` (i.e. `science score` y `history score`). Por lo tanto, pasamos de 11 + 2 = 13 columnas.
- Le칠mos `left` como el dataframe en la primera posici칩n de la funci칩n
`merge(LEFT DATAFRAME, RIGHT DATAFRAME, ...)`.
- Al utilizar el argumento `how='left'` dentro de la funci칩n `merge()`, 
estamos garantizando que mantendremos solo las llaves encontradas en
el DataFrame de la izquierda (`nombres`). Esto significa que solo se 
agregaran las columnas donde el nombre se encuentre en la izquierda, cualquier
nombre que este en el DataFrame `other_grades`, pero no en `df_grades`, su informaci칩n no ser치 anexada.
- Este m칠todo es relevante cuando queremos incrementar el n칰mero de informaci칩n
procedente de otras fuentes sobre un dominio de interes. Nos interesa
saber m치s de los alumnos que tienen todas las notas  `math score`, `reading score`, y `writing score`. Si tenemos notas de otras asignaturas, genial, pero
no queremos seguir alumnos que no tengan al menos las 3 primeras notas. Este
ser칤a un caso de uso para el `left join`.
- La llave al no ser especificada en `merge()` se determina por las columnas
que tienen igual nombre en ambos DataFrames (i.e. `nombre`).



````python
# Unir df_grades con other_grades.csv usando right join
df_right_join = pd.merge(
    df_grades,
    df_other,
    how="right"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_right_join: {df_right_join.shape}"
)
print(
    f"\nNulos df_right_join: \n{df_right_join.isna().sum()}"
)
df_right_join.head()
````

**Explicaci칩n**

- Similar al razonamiento de `how='left'`, pero con el DataFrame del segundo
argumento en `merge()`, el resultado de la operaci칩n contiene solo el n칰mero de filas originales de `other_grades`.
- Sin embargo, se modifica la informaci칩n a nivel de columnas, anexando las columnas adicionales que se encuentran en `df_grades` (i.e. `gender` a `simple parental level of education``). Por lo tanto, pasamos de 3 + 10 = 13 columnas.
- Le칠mos `right` como el DataFrame en la segunda posici칩n de la funci칩n
`merge(LEFT DATAFRAME, RIGHT DATAFRAME, ...)`.
- Al utilizar el argumento `how='right'` dentro de la funci칩n merge(), estamos garantizando que mantendremos solo las llaves encontradas en el DataFrame de la izquierda (`nombres`). Esto significa que solo se agregar치 las columnas donde el `nombre` se encuentre en la derecha, cualquier nombre que este en el DataFrame `df_grades`, pero no en `other_grades`, su informaci칩n no ser치 anexada.
- La llave al no ser especificada en `merge()` se determina por las columnas
que tienen igual nombre en ambos DataFrames (i.e. `nombre`).

````python
# Unir df_grades con other_grades.csv usando inner join
df_inner_join = pd.merge(
    df_grades,
    df_other,
    how="inner"
)
print(
    f"Dimensiones df_grades: {df_grades.shape}\n"
    f"Dimensiones df_other: {df_other.shape}\n"
    f"Dimensiones df_inner_join: {df_inner_join.shape}"
)
print(
    f"\nNulos df_inner_join: \n{df_inner_join.isna().sum()}"
)
df_inner_join.head()
````

**Explicaci칩n**

- Se anexa informaci칩n adicional para toda observaci칩n que contenga una
llave en el DataFrame `df_grades` Y `other_grades`. 
- Es una operaci칩n que se fija en la informaci칩n en la intersecci칩n de las
llaves de ambos DataFrame.
- La llave al no ser especificada en `merge()` se determina por las columnas que tienen igual nombre en ambos DataFrames (i.e. nombre).
- El DataFrame que da como resultado contiene 13 columnas, las 11 que se
encuentran en `df_grades` y dos nuevas columnas adicionales aportadas por
`other_grades`.
- Respecto al n칰mero de filas, tenemos un total de 875 observaciones, esto
quiere decir que hay solo 875 `nombres` compartidos en ambos DataFrames.
